```python
from util.hypertensi import age, sex, kolesterol, bmi, perokok, olahraga, kesehatan, konsumbuah, konsumsayur, ratekesehatan, diabetes
import pickle
import numpy as np
import sys

def hypertension_detect(input_text):
    
        input_text = input_text.lower().strip()
        if '#umur' in input_text:
            
            # input umur
            age_input = age(input_text)
            
            #  Input jenis kelamin
            sex_input = input('Apakah Anda seorang pria/wanita (#jk (pria/wanita))?')
            if '#jk' in sex_input:
                sex_input=sex(sex_input)
            else:  
                  return "inputan tidak sesuai. mohon masukkan kata '#jk (jenis-kelamin)'."
            
            # input penyakit kolesterol
            kolesterol_input = input('Apakah kolestrol Anda tinggi (#kolesterol (iya/tidak))?')
            if '#kolesterol' in kolesterol_input:
                kolesterol_input = kolesterol(kolesterol_input)
            else:
                  return "inputan tidak sesuai. mohon masukkan kata '#kolesterol (iya/tidak)'."
                  
            # input tinggi badan dan berat badan untuk BMI   
            bmi_input = input('Sebutkan tinggi dan berat badan Anda saat ini (#tb (tinggi badan) #bb (berat badan))?')
            if '#tb' in bmi_input and '#bb' in bmi_input:
                bmi_input = bmi(bmi_input)
            else:
                  return "inputan tidak sesuai. mohon masukkan kata '#jk (jenis-kelamin)'."
            
            # input apakah perokok atau tidak  
            perokok_input = input('Apakah Anda seorang perokok (#rokok (iya/tidak))?')
            if '#rokok'  in perokok_input:
                perokok_input = perokok(perokok_input)
            else:
                  return "inputan tidak sesuai. mohon masukkan kata '#rokok (iya/tidak)'."
                  
            
            # input apakah berolahraga
            olahraga_input = input('Apakah Anda berolahraga atau melakukan aktivitas selama 30 hari terakhir selain bekerja (#olahraga (iya/tidak))?')
            if '#olahraga' in olahraga_input:
                olahraga_input = olahraga(olahraga_input)
            else:
                  return "inputan tidak sesuai. mohon masukkan kata '(#olahraga (iya/tidak)'."
                  
            # input kesehatan   
            kesehatan_input = input('Apakah Anda mengalami penyakit selama 30 hari terakhir (#kesehatan (0-30))?')
            if '#kesehatan' in kesehatan_input:
                kesehatan_input = kesehatan(kesehatan_input)
            else:
                  return "inputan tidak sesuai. mohon masukkan kata '#kesehatan (0-30)'."
                  
            # input apakah mengkonsumsi buah satu hari sekali atau tidak  
            konsumbuah_input =input('Apakah Anda mengkonsumsi buah setidaknya satu kali sehari (#konsumbuah (iya/tidak))?')
            if '#konsumbuah'  in konsumbuah_input:
                konsumbuah_input = konsumbuah(konsumbuah_input)
            else:
                  return "inputan tidak sesuai. mohon masukkan kata '#konsumbuah (iya/tidak)'."
                  
            konsumsayur_input = input('Apakah Anda mengkonsumsi sayur setidaknya satu kali sehari (#konsumsayur (iya/tidak))?')
            if '#konsumsayur'  in konsumsayur_input:
                konsumsayur_input = konsumsayur(konsumsayur_input)
            else:
                  return "inputan tidak sesuai. mohon masukkan kata '#konsumsayur (iya/tidak)'."
                  
            ratekesehatan_input = input('Berapa rate atau persentase kesehatan Anda saat ini (#rtkesehatan (1-5))?')
            if '#rtkesehatan'  in ratekesehatan_input:
                ratekesehatan_input = ratekesehatan(ratekesehatan_input)
            else:
                  return "inputan tidak sesuai. mohon masukkan kata '#rtkesehatan (1-5)'."
                  
            diabetes_input = input('Apakah Anda pernah atau sedang menderita diabetes (#db (ya/tidak))?')
            if '#db'  in diabetes_input:
                diabetes_input = diabetes(diabetes_input)
            else:
                  return "inputan tidak sesuai. mohon masukkan kata '#db (ya/tidak)'."
                  
            # Load model dan membuat prediksi
            filename = 'save_data/hipertensi_prediction.pkl'
            gsnb = pickle.load(open(filename, 'rb'))
            cek_hipertensi = np.array([[age_input, sex_input, kolesterol_input, bmi_input, perokok_input,
                                        olahraga_input, kesehatan_input, konsumbuah_input, konsumsayur_input,
                                        ratekesehatan_input, diabetes_input]])
            prediction = gsnb.predict(cek_hipertensi)
            
            # Mengembalikan hasil prediksi
            if prediction == 1:
                return "Anda memiliki penyakit hipertensi. Jika Anda memiliki pertanyaan lebih lanjut, silakan tanyakan."
            else:
                return "Anda tidak memiliki penyakit hipertensi. Jika Anda memiliki pertanyaan lebih lanjut, silakan tanyakan."
        else:
            return "Input tidak sesuai dengan format. Mohon masukkan informasi umur dengan benar (gunakan format '#umur angka')."
```

```python

from util.hipertensi import age, bmi, diabetes, jkelamin, kesehatan, kolesterol, konsumbuah, konsumsayur, olahraga, perokok, ratekesehatan
import pickle
import pandas as pd
import numpy as np

# Load model dan membuat prediksi
filename = 'savedata/hipertensi_prection.pkl'
gsnb = pickle.load(open(filename, 'rb'))
            
def cek_hipertensi(input_text):
    input_text = input_text.lower()

    if '#cek' in input_text:

        hasil = [word for word in input_text.split() if not word.startswith("#")]
        if len(hasil) == 12:
            # Menghasilkan hasil dalam variabel terpisah
            age_value = hasil[0]
            Sex_value = hasil[1]
            HighChol_value = hasil[2]
            tinggi_value = hasil[3]
            berat_value = hasil[4]
            Smoker_value = hasil[5]
            PhysActivity_value = hasil[6]
            PhysHlth_value = hasil[7]
            Fruits_value = hasil[8]
            Veggies_value = hasil[9]
            GenHlth_value = hasil[10]
            Diabetes_value = hasil[11]
            
            # pemrosesan data
            age_input = age(age_value)
            sex_input = jkelamin(Sex_value)
            HighChol_input = kolesterol(HighChol_value)
            bmi_input = bmi(tinggi_value, berat_value)
            Smoker_input = perokok(Smoker_value)
            PhysActivity_input = olahraga(PhysActivity_value)
            PhysHlth_input = kesehatan(PhysHlth_value)
            Fruits_input = konsumbuah(Fruits_value)
            Veggies_input = konsumsayur(Veggies_value)
            GenHlth_input = ratekesehatan(GenHlth_value)
            Diabetes_input = diabetes(Diabetes_value)

            # Memasukkan data ke dalam DataFrame dengan nama fitur yang valid
            cek_hipertensi = pd.DataFrame([[age_input, sex_input, HighChol_input, bmi_input, Smoker_input,
                                PhysActivity_input, PhysHlth_input, Fruits_input, Veggies_input,
                                GenHlth_input, Diabetes_input]],
                                columns=['Age', 'Sex', 'HighChol', 'BMI', 'Smoker', 'PhysActivity',
                                        'PhysHlth', 'Fruits', 'Veggies', 'GenHlth', 'Diabetes'])
            prediction = gsnb.predict(cek_hipertensi)
            predict = prediction[0]

            # Mengembalikan hasil prediksi
            if predict == 1:
                return "Anda memiliki penyakit hipertensi. Jika Anda memiliki pertanyaan lebih lanjut, silakan tanyakan."
            else:
                return "Anda tidak memiliki penyakit hipertensi. Jika Anda memiliki pertanyaan lebih lanjut, silakan tanyakan."
        else:
            return 'data yang anda masukkan tidak lengkap'
    else:
        return "pesan yang kamu masukkan tidak sesuai format, sertakan '#cek'"
```

```python


tes = '#cek 25 pria tidak 165 65 tidak iya 0 tidak tidak 2 tidak'
cek = cek_hipertensi(tes)
print(cek)
type(cek)
```
